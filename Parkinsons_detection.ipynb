```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.preprocessing import StandardScaler
```

```python
parkinson_data=pd.read_csv('parkinsons.csv')
```

```python
data = parkinson_data.select_dtypes(include='number')
#only selecting numerical columns
```

```python
data.groupby('status').mean()
```

```python
X= data.drop(columns=['status'], axis=1)
```

```python
Y=data['status']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

```python
Model=svm.SVC(kernel='linear')
```

```python
Model.fit(X_train,Y_train)
```

```python
X_train_prediction= Model.predict(X_train)
Accuracy_on_data=accuracy_score(Y_train, X_train_prediction)
print(Accuracy_on_data)
```

```python
X_test_prediction= Model.predict(X_test)
Accuracy_on_data=accuracy_score(Y_test, X_test_prediction)
print(Accuracy_on_data)
```
